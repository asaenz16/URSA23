## The Classical Case 
In the classical case, the probability of finding the coin evens out over all possible positions as time gets very large. To run a simulation and test this, the first thing we need to do is some setup. We first define the probability as a function of the given position (x), current time (t), and length of ring (l). At time t = 0, we know that the coin is at x = 0 (this is our initial value).

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import seaborn
seaborn.set_theme()

# defines the probability of as a function of position (x),
# current time (t), and total length of the ring (l)
def prob_fun(x, t, l):
    values =[]
    temp = 0
    for k in range(l):
        temp = (1/l)*np.exp(2 *np.pi*(x-l)*(k/l)*(1j) + t*np.cos(2*np.pi*(k/l)) -t)
        temp = temp.real
        values. append(temp)
    temp = sum(values)
    return temp
```

Now we can generate values for $l = 10$ and $t = 100$. By inspection, we can see that these values are all pretty close together (all right around 0.1, which is $\frac{1}{l}$). As a sanity check, we can also check that the sum of all of the probabilities is 1.

```python
l=10
t = 100
vals =[]
for x in range(l):
    vals.append(prob_fun(x, t, l))
print(vals)
print(sum(vals))
```

Let's plot the probabilities against time and position. For this, we'll use $l = 20$. 

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

l = 20

# generate evenly spaced values for x and t
x = np.arange(-l+1, l+2, 1)
t = np.arange(0, l+2, 0.01)


X, T = np.meshgrid(x, t)
Z = prob_fun(X, T, 2*l +1)

surf = ax.plot_surface(X, T, Z, cmap = plt.cm.cividis)

# set labels for axes 
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('t', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

ax.view_init(25, 35)

plt.show()
```
