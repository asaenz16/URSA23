```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

def prob_fun(x, t, l):
    values =[]
    temp = 0
    for k in range(l):
        temp = (1/l)*np.exp(2 *np.pi*(x-l)*(k/l)*(1j) + t*np.cos(2*np.pi*(k/l)) -t)
        temp = temp.real
        values. append(temp)
    temp = sum(values)
    return temp
```

```python
l=10
t = 100
vals =[]
for x in range(l):
    vals.append(prob_fun(x, t, l))
print(vals)
print(sum(vals))
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

l = 20

x = np.arange(-l+1, l+2, 1)
t = np.arange(0, l+2, 0.01)


X, T = np.meshgrid(x, t)
Z = prob_fun(X, T, 2*l +1)

surf = ax.plot_surface(X, T, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('t', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

ax.view_init(25, 35)

plt.show()
```
